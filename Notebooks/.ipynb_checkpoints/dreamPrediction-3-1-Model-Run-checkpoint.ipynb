# Dream data (256 EEG times series at 500 Hz)

```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.io as sio
from sklearn.model_selection import train_test_split
from keras.utils import np_utils
```

```python
trial = (sio.loadmat('./data/JL02trial_1.mat')['a_trial']).T
```

```python
len(trial)
```

```python
plt.figure(figsize=(20,10))

for i in range(0,256, 2):
    plt.plot(trial[:1000,i])

plt.show()
```

```python
# End of the time series (last 
plt.figure(figsize=(20,10))

for i in range(0,256, 2):
    plt.plot(trial[-1000:,i])

plt.show()
```

```python
plt.plot(trial[0])
plt.plot(trial[30000])
plt.plot(trial[-1])
```

```python
plt.plot(trial[0])
plt.plot(trial[500])
plt.plot(trial[1000])
```

```python
num_steps = 1000

X = np.concatenate((trial[:num_steps], trial[30000:30000+num_steps], trial[-num_steps:]), axis=0)[:,:256]
y = [0]*num_steps + [1]*num_steps + [2]*num_steps
```

```python
X.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# encode labels as one-hot vectors
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)
```

```python
y_test[:10]
```

```python
import tensorflow as tf 
```

```python
from keras.models import Sequential
from keras.layers import Dense
from keras import regularizers
from keras import optimizers
from keras import initializers
```

```python
# ONE HIDDEN LAYER:

# define model  (see Michael Nielsen, "NN and DL")
model = Sequential()
# first layer
model.add(Dense(100,   # or 100
                input_dim=256, 
                kernel_initializer='normal',   # 'normal', initializers.Constant(value=0), ...
#                 kernel_regularizer=regularizers.l2(0.01),  # smooth filters, but bad accuracy
                activation='sigmoid'))  # 'relu', 'sigmoid', 'tanh', ...
# second layer
#model.add(Activation(tf.nn.softmax))
model.add(Dense(3, 
                kernel_initializer='normal',   # 'normal', ...
#                 kernel_regularizer=regularizers.l2(0.1),  # smooth filters, but bad accuracy
                activation='softmax'))

# compile
model.compile(loss='categorical_crossentropy',
              optimizer='adam', # 'adam', optimizers.SGD(lr=0.1), ...
              metrics=['accuracy'])
```

```python
# fit and evaluate model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=50, batch_size=10, verbose=2)
scores = model.evaluate(X_test, y_test, verbose=0)
print("Baseline Error: %.2f%%" % (100-scores[1]*100))
```

# CNN with keras (for 16x16 EEG images)

```python
X_train.shape
```

```python
# reshape MNIST data to fit model
X_train = X_train.reshape(X_train.shape[0],16,16,1)
X_test = X_test.reshape(X_test.shape[0],16,16,1)
```

```python
X_train.shape
```

```python
plt.matshow(X_train[5].reshape(16,16))
```

```python
from keras.models import Sequential
from keras.layers import Dense, Conv2D, Flatten
# create model
model = Sequential()
# add layers
model.add(Conv2D(64, kernel_size=3, activation='relu', input_shape=(16,16,1)))
model.add(Conv2D(32, kernel_size=3, activation='relu'))
model.add(Flatten())
model.add(Dense(3, activation='softmax'))
```

```python
# compile model
model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# train the model
model.fit(X_train, y_train, validation_data=(X_test, y_test), epochs=10)
```
